## Feature extraction and preprocessing

```python
x_train=train_data.drop(columns='target')
y_train=train_data['target']
print(x_train)
print(y_train) 
```

```python
x_test=test_data.drop(columns='target')
y_test=test_data['target']
print(x_test)
print(y_test)
```

```python
cat_columns = ['cp', 'exang', 'slope', 'thal']
num_columns = [c for c in x_train.columns if c not in cat_columns]
```

```python
num_columns
```

## One-Hot Encoding in train data

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#create an oneHotEncoder instance
# transform on training data
encoder=OneHotEncoder(handle_unknown='ignore')
#fit on categorical column
encoder.fit(x_train[cat_columns])
#columns name
column_names=encoder.get_feature_names(input_features=cat_columns)
# transform on training data
x_train_cat_encoded=encoder.transform(x_train[cat_columns])



x_train_cat_encoded_df=pd.DataFrame(x_train_cat_encoded.todense(),
                                    columns=column_names,
                                    index=x_train.index)
x_train_cat_encoded_df
```

## Final training set

```python
x_train_encoded=pd.concat([x_train[num_columns],x_train_cat_encoded_df],axis=1)
x_train_encoded
```

## One-Hot Encoding in test data

```python
# transform on testing data
x_test_cat_encoded=encoder.transform(x_test[cat_columns])



x_test_cat_encoded_df=pd.DataFrame(x_test_cat_encoded.todense(),
                                    columns=column_names,
                                    index=x_test.index)
x_test_encoded=pd.concat([x_test[num_columns],x_test_cat_encoded_df],axis=1)
print(x_test_encoded)
```

### Model Building and Grid Search

## SVM

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_param1 = {'kernel':('linear', 'rbf'), 'C':(1, 10)}
```

```python
from sklearn.metrics import make_scorer, f1_score
from sklearn.svm import SVC
```

```python
scorer = make_scorer(f1_score, average='micro')
clf = GridSearchCV(SVC(), grid_param1,scoring=scorer,cv=10)
clf.fit(x_train_encoded, y_train)
```

```python
print(clf.best_score_, clf.best_params_)
```

## Model Evalution

```python
import sklearn.metrics as metrics

model = SVC(random_state=1,kernel='linear', C=10)
model.fit(x_train_encoded, y_train)
y_pred = model.predict(x_test_encoded)
print(metrics.classification_report(y_test, y_pred))
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, f1_score# error something score
```

```python

grid_param2 = {
    'max_depth': (3, 5, 7, 9, 11, 13),
    'min_samples_split': (2, 4, 6, 8, 10)
}
```

```python
scorer = make_scorer(f1_score, average='micro')
clf=GridSearchCV(DecisionTreeClassifier(),grid_param2,scoring=scorer,cv=10)
clf.fit(x_train_encoded, y_train)
```

```python
print(clf.best_score_, clf.best_params_)
```

## Model Evalution

```python

import sklearn.metrics as metrics
model =DecisionTreeClassifier(random_state=1,max_depth=3,min_samples_split=4)
model.fit(x_train_encoded, y_train)
preds = model.predict(x_test_encoded)
print(metrics.classification_report(y_test, preds))
```
